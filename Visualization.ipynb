```python
import pandas as pd
import numpy as np
import csv
import os
GAME_FILE = 'mlb_data/games.csv'
```

```python
data = {'2016': dict(), '2017': dict(), '2018': dict(), '2019': dict(), '2020': dict(), '2021': dict()}
keyword = ['WILD', 'ALDS', 'NLDS', 'ALCS', 'NLDS', 'WORLD SERIES']
```

```python
raw_data = pd.read_csv(GAME_FILE)
game_id = list()
for i in range(len(raw_data)):
    if not pd.isnull(raw_data['postseason info'][i]):
        print('not consider postseason', raw_data['home'][i], raw_data['away'][i], raw_data['Date'][i])
        continue
    if raw_data['Duration'][i] == 'Canceled':
        print('the game is canceled', )
        continue
    if raw_data['Game'][i] in game_id:
        print('Duplicate Game!')
        continue
    
    # save the game id
    game_id.append(raw_data['Game'][i])
    
    away = raw_data['away'][i]
    home = raw_data['home'][i]
    season = raw_data['Date'][i][0:4]
    print(i, season, away, home)
    if away not in data[season]:
        data[season][away] = {'score': 0, 'walks_issued': 0, 'total_bases': 0, 'strikeouts': 0, 'stolen_bases': 0, 'game_played': 0}
    if home not in data[season]:
        data[season][home] = {'score': 0, 'walks_issued': 0, 'total_bases': 0, 'strikeouts': 0, 'stolen_bases': 0, 'game_played': 0}
    # update score
    data[season][away]['score'] += int(raw_data['away-score'][i])
    data[season][home]['score'] += int(raw_data['home-score'][i])
    
    # update walk
    data[season][away]['walks_issued'] += int(raw_data['Walks Issued - Away'][i])
    data[season][home]['walks_issued'] += int(raw_data['Walks Issued - Home'][i])
    
    # update total bases
    data[season][away]['total_bases'] += int(raw_data['Total Bases - Away'][i])
    data[season][home]['total_bases'] += int(raw_data['Total Bases - Home'][i])
    
    # update stikeout
    data[season][away]['strikeouts'] += int(raw_data['Strikeouts Thrown - Away'][i])
    data[season][home]['strikeouts'] += int(raw_data['Strikeouts Thrown - Home'][i])
    
    # update stolen bases
    data[season][away]['stolen_bases'] += int(raw_data['Stolen Bases - Away'][i])
    data[season][home]['stolen_bases'] += int(raw_data['Stolen Bases - Home'][i])
    
    # update game play game_played
    data[season][away]['game_played'] += 1
    data[season][home]['game_played'] += 1
```

```python
csv_columns = ['team', 'season', 'score', 'walks_issued', 'total_bases', 'strikeouts', 'stolen_bases', 'game_played']

dict_data = list()
for key in data:
    for key1 in data[key]:
        if key1 == 'AL' or key1 == 'NL':
            continue
        t = dict()
        t['team'] = key1
        t['season'] = key
        t['score'] = data[key][key1]['score']
        t['walks_issued'] = data[key][key1]['walks_issued']
        t['total_bases'] = data[key][key1]['total_bases']
        t['strikeouts'] = data[key][key1]['strikeouts']
        t['stolen_bases'] = data[key][key1]['stolen_bases']
        t['game_played'] = data[key][key1]['game_played']
        dict_data.append(t)
    
csv_file = "data.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for d in dict_data:
            writer.writerow(d)
except IOError:
    print("I/O error")



```
